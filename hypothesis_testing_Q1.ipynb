Q1.A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.

Answer: We are going to conduct a 2 tailed t-Test on 2 Independent samples with Numerical Data.

We need to check whether the mean of both samples are different and
Is there any significance difference between the two samples?

Step 1.

Null Hypothesis: There is no significant difference between the diameters of cutlet between two units. u1 = u2
   

Alternate Hypothesis: There is significant difference between the diameters of cutlet from two units. 
                    u1 not equal to u2

**Step 2:**
Decide a cut-off value

*   Significance 5%
*   alpha = 0.05

As it is a two-tailed test

*  alpha/2 = 0.025





**Step 3**

```python
import scipy.stats as stats
import statsmodels.api as sm
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
cutlets = pd.read_csv('Cutlets.csv')
cutlets.head(10)
```

```python
statistic , p_value = stats.ttest_ind(cutlets['Unit A'],cutlets['Unit B'], alternative = 'two-sided')
print('p_value=',p_value)
```

**Step 4**
Compare p_value with '
 '(Significane Level)
If p_value is <= alpha: we reject Null Hypothesis because of lack of evidence

If p_value is > alpha we  failed to  reject Null Hypothesis

interpreting p-value

```python
alpha = 0.025
print('Significnace=%.3f, p=%.3f' % (alpha, p_value))
if p_value <= alpha:
    print('We reject Null Hypothesis there is a significance difference between two Units A and B')
else:
    print('We fail to reject Null hypothesis')
```

Hence, We fail to reject Null Hypothesis because of lack of evidence, there is no significant difference between the two samples
